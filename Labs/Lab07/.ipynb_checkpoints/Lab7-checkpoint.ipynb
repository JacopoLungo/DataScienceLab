```python
from os.path import dirname, join as pjoin
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from os import listdir
import re
from IPython.display import Audio
```

```python
from scipy.fftpack import fft
```

```python
#function to plot the audio wave given a array of numbers
def graphAudio(data, samplerate = 8000):
    plt.rcParams['figure.figsize'] = [10, 5]
    length = data.shape[0] / samplerate
    time = np.linspace(0., length, data.shape[0])
    plt.plot(time, data)
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude")
    plt.show()
```

```python
#function to listent to audios. data param is an array of numbers
def listenAudio(data, samplerate = 8000):
    return Audio(data = data, rate = samplerate)
```

```python
# store all the datas and all the lables
X = []
y = []
for file in listdir(".\data\dev"):
    result = re.search(r"_\w", file)
    y.append(int(result.group(0)[1]))
    X.append(wavfile.read("./data/dev/" + file)[1])
X = np.array(X, dtype = object)
y = np.array(y)
```

```python
# Check that all the classes are equally represented
pd.Series(y).value_counts()
```

```python
graphAudio(fft(X[0]))
```

```python
from scipy.io import wavfile
from scipy.signal import stft

# Read the WAV file
sample_rate, audio = wavfile.read("0_4.wav")

# Compute the STFT of the audio signal
f, t, Zxx = stft(audio, sample_rate, nperseg=1024)

# Plot the spectrogram
plt.pcolormesh(t, f, np.abs(Zxx))
plt.title('Spectrogram')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
from scipy.signal import stft

# Use the stft() function to compute the short-time Fourier transform of the audio data
f, t, Zxx = stft( X[0] )

# Use the imshow() function to plot the spectrogram
plt.imshow(Zxx, extent=[t.min(), t.max(), f.min(), f.max()])
plt.title("Spectrogram")
plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")
plt.show()
```
